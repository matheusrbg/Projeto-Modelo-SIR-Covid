###**Resumo** 

Em 18 de Janeiro de 2021 foram iniciadas as vacinações no estado de Pernambuco contra a Covid-19. Por isso, se torna bastante interessante desenvolvermos modelos que consigam estimar a evolução da doença diante o novo cenário, com a presença das vacinas, o qual possui grande benefícios no controle da pandemia.

Pretendemos analisar 3 casos. O primeiro caso seria a evolução da doença na taxa atual ao que ela se encontra com a presença da vacinação, o segundo caso seria a análise da doença se não houvesse vacina e o terceiro caso seria se houvesse um sistema de vacinação mais eficiente, fazendo com que mais indivíduos consigam ser vacinados, diminuindo a taxa de infectados.

Para realizar essas análises, utilizaremos um modelo variante do SIR, o SIRV, o qual analisará o número de indivíduos suscetíveis, infectados, removidos e vacinados, para estimar a evolução da doença com e sem as vacinas, assim como a progressão da taxa de vacinados da população no território de Pernambuco. Buscando assim, evidenciar que as vacinas, mesmo que não sejam 100% eficazes, são de vital importância para a mitigação dos transtornos causados pelo vírus.

###**Introdução e objetivos**

####Introdução:


Recentemente, em dezembro de 2019, a cidade de Wuhan, na China, se tornou o centro de um surto de uma “pneumonia” de origem desconhecida. Em janeiro os cientistas chineses descobriram a origem de um novo coronavírus (SARS-CoV-2) o qual estaria causando essas pneumonias. Em fevereiro de 2020 essas pneumonias foram nomeadas, pela World Health Organization, de Covid-19, uma nova doença que se alastrou pelo mundo inteiro, se tornando não somente um surto mas uma pandemia.

Na busca de entender melhor essa nova doença, foram realizadas várias pesquisas
científicas e análises estatísticas buscando encontrar modelos que consigam predizer como a doença irá se comportar em diferentes regiões e populações.

Pretendemos analisar a taxa de evolução da doença no estado de Pernambuco, Brasil, assim como predizer o período que será o pico de infectados, além estimar a quantidade de mortes causadas pela doença e analisar a progressão da vacinação. Optamos por utilizar um modelo variante do original SIR, chamado de SIRV (Susceptible-Infected-Removed-Vaccinated), que levará em consideração o caso de pessoas vacinadas, para estimarmos com mais fidelidade o avanço da Covid-19.

####Objetivos:

Tendo em mente que uma parte da população ainda tem dúvidas equivocadas sobre a necessidade e a eficácia da vacinação, esse projeto terá como objetivo fazer o contraste do cenário atual com o do desenvolvimento do COVID-19 caso não houvessem vacinas. Além disso, analisaremos o gráfico no caso de uma taxa de vacinação maior. Reforçando, assim, a necessidade de tomar a vacina e derrubando o mito que a vacina é ineficaz. 

###**Métodos**

O modelo matemático SIR (suscetível, infectados e removidos) é o mais simples para simular uma pandemia, nosso projeto utilizará uma variante desse modelo: O SIRV. Essa variante também leva em consideração a população dos vacinados. Assumindo que a população de Pernambuco permanece constante com o tempo, temos as seguintes equações:

$N=S(t)+I(t)+R(t)+V(t)$
 
$0=S′+I′+R′+V′$
 
$\frac{dS(t)}{dt} =−\frac{β}{N}S(t)I(t)−αS(t)$
 
$\frac{dI(t)}{dt} =\frac{β}{N}S(t)I(t)−(γ+µ)I(t)$
 
$\frac{dR(t)}{dt} =γI(t)- αR(t)(1-µ)$
 
$\frac{dV(t)}{dt} =αS(t) + αR(t)(1-µ)$

$\frac{dD(t)}{dt}= µI(t)$

Onde:

* N é a população total
* S(t): representa a população suscetível a ter a doença no tempo t, ou seja, é a população que não foi infectada e que não possui imunização.
* I(t): representa a população infectada com o vírus no tempo t, esses são os indivíduos que estão infectados e podem espalhar a doença para a população que está no estado suscetível.
* R(t): representa a população removida no tempo t, dizer que uma população foi removida nesse modelo significa que ou ficou recuperada da doença.
* V(t): representa a população efetivamente vacinada no tempo t. São os indivíduos que foram vacinados e ganharam imunidade total à doença.
* D(t): é a quantidade de mortos pelo Covid-19.
* β: Representa a taxa de transmissão por dia da doença, o qual é calculado sendo a quantidade de contatos que um indivíduo tem com outros vezes a probabilidade de transmissão de um indivíduo infectado com outro suscetível. Tal parâmetro pode ser modificado utilizando técnicas de combate a pandemia como o distanciamento social
* γ: Representa a taxa de recuperação dos indivíduos infectados por dia, a taxa 1/γ representa a média do período de infecção da doença.
* α: Representa a taxa de vacinação e a eficácia da vacina, é calculada fazendo a fração de indivíduos que ganharam imunização total com todos os que foram vacinados multiplicado pela taxa de vacinação de pernambuco.
* µ: Taxa de mortalidade por dia do Covid-19

Estimativa e coleta de dados:

Pegamos dados do IBGE para estimar a população de Pernambuco no início do ano de 2021. Com isso temos N = 9616621.

Para o número de infectados iniciais nós pegamos os dados de “JHU CSSE COVID-19 Data”, o qual realizam uma média entre os 7 dias, por isso multiplicamos por 7, resultando em I0 = (1236)*7

Como estamos analisando no período de início da vacinação temos que o V0 = 0

Para calcularmos os Removidos iniciais, estamos somando a quantidade de mortos e casos confirmados do dia 18 de janeiro em 2021 de Pernambuco.

Com isso, temos o número de suscetíveis sendo o total da população subtraído com os  removidos iniciais, vacinados iniciais e infectados iniciais.

Temos que a taxa de transmissão é aproximadamente igual a β = 0,17, pois temos dados que aproximam o R0 a 2,5, e sabendo que R0 = β/γ chegamos a esse valor. 

Como sabemos que o período médio de infecção da Covid- 19 é 14 dias temos que γ = 1/14 = 0,07. 

Para calcularmos a taxa de eficácia da vacina utilizei a média das eficácias gerais de 3 vacinas que estão sendo administradas no território de Pernambuco (Pfizer com 95% de eficácia geral, Astrazeneca com 82,4% de eficácia geral e CoronaVac 50,38% de eficácia geral) resultando em 0,759.

Para calcularmos a taxa de vacinação do estado de Pernambuco dividimos a média de pessoas vacinadas por dia no estado de Pernambuco pela população total.

Para calcularmos a porcentagem de pessoas imunizadas por dia, o α, nós multiplicamos a taxa de vacinação do estado de Pernambuco pela eficácia da vacina. 

A taxa de mortalidade utilizamos os dados do site oficial de monitoramento de casos do Brasil e fizemos a fração (óbitos)/(casos confirmados). Resultando em uma taxa de 0,031 mortos por quantidade de contaminados desde o começo da doença, dividindo isso pela quantidade de dias desde 18 de janeiro de 2021,temos µ = 0,031/192 como o número aproximado da taxa de pacientes infectados mortos por dia.

### Dados, modelos e bibliotecas comuns às simulações

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#Modelos:

# SIR Model
def deriv_sir(y, t):
    S, I, R, D = y
    dSdt = (-beta * S * I / N)
    dIdt = (beta * S * I / N) - ((gamma+mi) * I)
    dRdt = (gamma * I)
    dDdt = I*mi
    return dSdt, dIdt, dRdt, dDdt

# SIRV Model
def deriv_sirv(y, t):
    S, I, R, V, D = y
    dSdt = (-beta * S * I / N) - (alfa*S)
    dIdt = (beta * S * I / N) - ((gamma+mi) * I)
    dRdt = (gamma * I) - (alfa*R*(1.-mi))
    dVdt = (alfa*S) + (alfa*R*(1.-mi))
    dDdt = I*mi
    return dSdt, dIdt, dRdt, dVdt, dDdt
```

```python
#Dados:

# Total da população estimada de Pernambuco em 18 de Janeiro de 2021
N = 9616621
# Número de infectados de Pernambuco em 18 de Janeiro de 2021 (Utilizando uma média) (Condições iniciais)
I0 = (1236)*7
# Números de vacinados de Pernambuco em 18 de Janeiro de 2021 (Condições iniciais)
V0 = 0
# Números de removidos de Pernambuco em 18 de Janeiro de 2021 (Condições iniciais)
R0 = (213329) + (10031)
# Número de novos óbitos em Pernambuco a partir de 18 de Janeiro de 2021 (Condições iniciais)
D0 = 0
# Número total de suscetíveis em Pernambuco em 18 de Janeiro de 2021 (Condições iniciais)
S0 = N - I0 - V0 - R0
# Taxa de contaminação por dia
beta = 0.17
# Taxa de recuperação (1/dias de infecção)
gamma = 1/14
# Taxa de mortalidade por dia do Covid-19
mi = 0.031/192

# Análise do gráfico em 200 dias
t = np.linspace(0, 200, 200)
```

### **Análise das condições atuais no estado de Pernambuco**

```python
# Taxa de vacinação eficaz por dia em condições atuais  
alfa = 0.006*0.759

# Vetor de condições iniciais
y0 = S0, I0, R0, V0, D0

# Integração das equações SIRV
ret = odeint(deriv_sirv, y0, t)
S_atual, I_atual, R_atual, V_atual, D_atual = ret.T

#Criando 2 figuras para os graficos 
fig, axs = plt.subplots(1, 2, figsize=(12, 5), constrained_layout=True,
                        sharex=True, sharey=False, facecolor='#dddddd')

# Plot the data of S(t), I(t), R(t), V(t)
axs.flat[0].grid(b=True, which='major', c='w', lw=1, ls='-')
axs.flat[0].plot(t, S_atual/N, 'b', alpha=0.5, lw=2, label='Susceptible')
axs.flat[0].plot(t, I_atual/N, 'r', alpha=0.5, lw=2, label='Infected')
axs.flat[0].plot(t, R_atual/N, 'g:', alpha=0.5, lw=2, label='Recovered')
axs.flat[0].plot(t, V_atual/N, 'y-.', alpha=0.5, lw=2, label='Vaccinated')
axs.flat[0].legend(['Susceptible', 'Infected','Recovered','Vaccinated'])
axs.flat[0].set_xlabel('Time in days since January 18, 2021')
axs.flat[0].set_ylabel('Population(%)')
axs.flat[0].grid(b=True, which='major', c='#dddddd', lw=2, ls='-')
axs.flat[0].set_xlim(0,200)
axs.flat[0].yaxis.set_tick_params(length=0)  # Não aparecer números quebrados
axs.flat[0].xaxis.set_tick_params(length=0)
axs.flat[0].set_title('SIRV')


#Plot the data of D(t) in a bar plot
axs.flat[1].fill_between(t , 0, D_atual, facecolor='black', alpha=0.5)
axs.flat[1].legend(['Dead'], loc = 2)
axs.flat[1].set_xlabel('Time in days since January 18, 2021')
axs.flat[1].set_ylabel('Deaths due to Covid-19 since January 18, 2021')
axs.flat[1].set_ylim(0,18000)
axs.flat[1].yaxis.set_tick_params(length=0)  # Não aparecer números quebrados
axs.flat[1].xaxis.set_tick_params(length=0)
axs.flat[1].set_title('Mortos')

# Números finais 
print('Mortes acumuladas por covid na simulação:') 
print(round(D_atual[199])) 
print('Vacinações acumuladas na simulação:')
print(round(V_atual[199]))
print('Recuperados que não foram vacinados acumulados na simulação:')
print(round(R_atual[199]))

plt.savefig('sirv.pdf')
plt.show()
```

### **Análise no estado de Pernambuco sem vacinação**

```python
#Taxa de Vacinação e quantidade de Vacinados seria igual a zero se as vacinações nunca houvessem começado

# Vetor de condições iniciais
y0 = S0, I0, R0, D0

# Integração das equações SIR
ret = odeint(deriv_sir, y0, t)
S_semVac, I_semVac, R_semVac, D_semVac = ret.T

#Criando 2 figuras para os graficos 
fig, axs = plt.subplots(1, 2, figsize=(12, 5), constrained_layout=True,
                        sharex=True, sharey=False, facecolor='#dddddd')

# Plot the data of S(t), I(t), R(t), V(t)
axs.flat[0].grid(b=True, which='major', c='w', lw=1, ls='-')
axs.flat[0].plot(t, S_semVac/N, 'b', alpha=0.5, lw=2, label='Susceptible')
axs.flat[0].plot(t, I_semVac/N, 'r', alpha=0.5, lw=2, label='Infected')
axs.flat[0].plot(t, R_semVac/N, 'g:', alpha=0.5, lw=2, label='Recovered')
axs.flat[0].legend(['Susceptible', 'Infected','Recovered'])
axs.flat[0].set_xlabel('Time in days since January 18, 2021')
axs.flat[0].set_ylabel('Population(%)')
axs.flat[0].grid(b=True, which='major', c='#dddddd', lw=2, ls='-')
axs.flat[0].set_xlim(0,200)
axs.flat[0].yaxis.set_tick_params(length=0)  # Não aparecer números quebrados
axs.flat[0].xaxis.set_tick_params(length=0)
axs.flat[0].set_title('SIR')


#Plot the data of D(t) in a bar plot
axs.flat[1].fill_between(t , 0, D_semVac, facecolor='black', alpha=0.5)
axs.flat[1].legend(['Dead'], loc = 2)
axs.flat[1].set_xlabel('Time in days since January 18, 2021')
axs.flat[1].set_ylabel('Deaths due to Covid-19 since January 18, 2021')
axs.flat[1].set_ylim(0,18000)
axs.flat[1].yaxis.set_tick_params(length=0)  # Não aparecer números quebrados
axs.flat[1].xaxis.set_tick_params(length=0)
axs.flat[1].set_title('Mortos')

# Números finais 
print('Mortes acumuladas por covid na simulação:')
print(round(D_semVac[199]))
print('Recuperados acumulados na simulação:')
print(round(R_semVac[199]))

plt.savefig('sir.pdf')
plt.show()
```

### **Resultados e discussões**


Podemos observar, como esperado, que o pico dos infectados no caso em que há uma vacina é menor e levemente deslocado para a direita em relação ao cenário sem vacina, o que mostra que conseguimos, através da vacina, adiar e, principalmente, controlar esse estouro de casos, fazendo com que menos pessoas sofram com as consequências dessa doença.

É possível observar também que, caso a campanha de vacinação estivesse sendo mais eficiente, esse pico dos infectados seria ainda menor e, consequentemente, o número de mortos também, comprovando que a vacina tem um impacto direto na progressão ou regressão da doença e que, por isso, devemos buscar ampliar a vacinação da população o máximo possível e incentivar todos que podem a se vacinarem. Além disso, vale frisar que nossa simulação, que começa em 18 de janeiro de 2021, é compativel com os dados atuais sobre a progressão da doença e conseguiu prever com uma precisão significante.

A quantidade de mortos em 2021, com a campanha de vacinação atual, calculada por nossas equações, foi cerca de 9,538 pessoas. De acordo com o gráfico oficial de óbitos por covid em Pernambuco [9], o número de óbitos foi de 8,833, obtemos um erro de 7.4%.

Oficialmente [12], temos 4,531,487 de pernambucanos com a primeira dose (dados de 08/08), nossas equações acharam o valor de 5,579,479 vacinados. Assim, nesse aspecto, obtemos um erro de 18.7%. Acreditamos que esse grande erro se deu porque ao calcularmos os recuperados nós subtraímos os vacinados, o que justifica essa disparidade entre os valores encontrados.

### **Conclusão**

Diante dos resultados obtidos do modelo SIRV percebemos que a doença possui uma taxa de infecção bastante elevada podendo ter consequências desastrosas. Já quando temos o cenário da vacinação, torna-se evidente um controle bastante significativo da doença, gerando um menor pico de infectados e uma redução no total de mortos em Pernambuco. Além disso, quanto mais eficaz for o projeto de vacinação do Estado mais essas reduções são observadas. Com isso podemos concluir que a vacinação é sim a maneira mais eficaz de se finalizar a pandemia do Covid-19 em Pernambuco e no Mundo.
